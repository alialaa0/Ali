# **Data Preprocessing**

```python
 import pandas as pd
 import numpy as np
```

```python
# Read CSV Data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/freelance yard/Sales Dashboard for E-commerce Business/01JCB2EBKMK3MX13K3NF0DQSBS.csv')
data
```

```python
data['Average Order Size'] = data['Total_Sales'] / data['Quantity_Sold']
```

```python
# Unnecessary columns with names like Unnamed: 15 are dropped.
columns_to_drop = ['Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17' , 'Unnamed: 18' ,'Unnamed: 19' , 'Unnamed: 20',
                   'Unnamed: 21' , 'Unnamed: 22' , 'Unnamed: 23' , 'Unnamed: 24' , 'Unnamed: 25' , 'Unnamed: 26' ,'Unnamed: 27',
                   'Unnamed: 28' , 'Unnamed: 29']  # Replace with actual column names

data = data.drop(columns=columns_to_drop)

data
```

```python
data.info()
```

```python
# The Time column is converted into a datetime
data['Time'] = pd.to_datetime(data['Time'], errors='coerce')
```

```python
data['Time']
```

```python
# The Time is categorized into periods of the day
data['Time of Day'] = pd.cut(data['Time'].dt.hour,bins=[0, 6, 12, 18, 24], labels=['Night', 'Morning', 'Afternoon', 'Evening'], right=False)
```

**Calculate KPIs (Key Performance Indicators)**

```python
# Calculate the total sales

total_sales = data['Total_Sales'].sum()
print(f"Total Sales: {total_sales}")
```

```python
# Calculate the total profit

total_profit = data['Profit'].sum()
print(f"Total Profit: {total_profit}")
```

```python
# Calculate the total quantity sold

total_quantity = data['Quantity_Sold'].sum()
print(f"Total Quantity Sold: {total_quantity}")
```

```python
# If the "Average Order Size" column already exists

average_order_size = data['Average Order Size'].mean()
print(f"Average Order Size: {average_order_size}")
```

```python
# If "Return_Flag" contains 1 for returned orders and 0 for others

return_rate = data['Return_Flag'].mean() * 100
print(f"Return Rate: {return_rate:.2f}%")
```

 **Calculate Month-over-Month and Year-over-Year Growth**

```python
# Group sales data by Year and Month and calculate total sales

monthly_sales = data.groupby(['Year', 'Month'])['Total_Sales'].sum().reset_index()
print(monthly_sales.head())
```

```python
# Calculate Month-over-Month Growth

monthly_sales['MoM Growth'] = monthly_sales['Total_Sales'].pct_change()
print(monthly_sales[['Year', 'Month', 'Total_Sales', 'MoM Growth']].head())
```

```python
# Group sales data by Year and calculate total sales
annual_sales = data.groupby('Year')['Total_Sales'].sum().reset_index()


# Calculate Year-over-Year Growth
annual_sales['YoY Growth'] = annual_sales['Total_Sales'].pct_change()
print(annual_sales)
```

**Identify Top Products, Regions, and Customers**

```python
# Group sales data by Product_ID and calculate total sales, then sort in descending order

top_products = data.groupby('Product_ID')['Total_Sales'].sum().sort_values(ascending=False).head(2)
print("Top Products by Total Sales:")
print(top_products)
```

```python
# Group sales data by Region_ID and calculate total sales, then sort in descending order

top_regions = data.groupby('Region_ID')['Total_Sales'].sum().sort_values(ascending=False).head(2)
print("Top Regions by Total Sales:")
print(top_regions)
```

```python
# Group sales data by Sales_Rep_ID and calculate total sales, then sort in descending order

top_customers = data.groupby('Sales_Rep_ID')['Total_Sales'].sum().sort_values(ascending=False).head(2)
print("Top Customers by Total Sales:")
print(top_customers)
```

**Analyze Sales by Time**

```python
# Group sales data by Day_of_Week and calculate total sales, then sort in descending order
sales_by_day = data.groupby('Day_of_Week')['Total_Sales'].sum().sort_values(ascending=False)
print("Sales by Day of the Week:")
print(sales_by_day)
```

```python
# Group sales data by Time of Day and calculate total sales, then sort in descending order
sales_by_time = data.groupby('Time of Day')['Total_Sales'].sum().sort_values(ascending=False)
print("Sales by Time of Day:")
print(sales_by_time)
```

```python
# Create a line plot for monthly sales trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Total_Sales', hue='Year', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Hourly Sales Distribution (Line Chart)

hourly_sales = data.groupby(data['Time'].dt.hour)['Total_Sales'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_sales.index, y=hourly_sales.values, marker='o')
plt.title('Hourly Sales Distribution')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Sample DataFrame (replace this with your actual dataset)
# data = pd.read_csv('sales_data.csv')  # Your actual dataset

# Ensure the 'Time' column is of datetime type
data['Time'] = pd.to_datetime(data['Time'], errors='coerce')  # Coerce invalid dates to NaT

# Example data for testing (you can replace it with your actual data)
data = pd.DataFrame({
    'Region_ID': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Sales_Rep_ID': ['Rep1', 'Rep2', 'Rep3', 'Rep1', 'Rep2', 'Rep3'],
    'Product_ID': ['Prod1', 'Prod2', 'Prod3', 'Prod1', 'Prod2', 'Prod3'],
    'Total_Sales': [1000, 1500, 2000, 2500, 1200, 1800],
    'Profit': [400, 600, 800, 1000, 500, 700],
    'Return_Rate': [5, 4, 3, 6, 2, 5],
    'Day_of_Week': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
    'Hour': [9, 10, 11, 12, 13, 14],
    'Time': pd.date_range('2024-01-01', periods=6, freq='D')
})

# Create a Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='region-filter',
        options=[{'label': region, 'value': region} for region in ['All Regions'] + data['Region_ID'].unique().tolist()],
        value='All Regions',  # Default value
        placeholder="Select Region"
    ),
    dcc.Dropdown(
        id='sales-rep-filter',
        options=[{'label': rep, 'value': rep} for rep in ['All Sales Reps'] + data['Sales_Rep_ID'].unique().tolist()],
        value='All Sales Reps',  # Default value
        placeholder="Select Sales Rep"
    ),
    dcc.Dropdown(
        id='product-filter',
        options=[{'label': product, 'value': product} for product in ['All Products'] + data['Product_ID'].unique().tolist()],
        value='All Products',  # Default value
        placeholder="Select Product"
    ),
    dcc.Graph(id='sales-by-day-graph'),
    dcc.Graph(id='sales-by-time-graph'),
    dcc.Graph(id='profit-vs-returns-graph'),
    dcc.Graph(id='product-sales-graph'),
    html.Div(id='observations-container')
])

@app.callback(
    Output('sales-by-day-graph', 'figure'),
    Output('sales-by-time-graph', 'figure'),
    Output('profit-vs-returns-graph', 'figure'),
    Output('product-sales-graph', 'figure'),
    Output('observations-container', 'children'),
    [Input('region-filter', 'value'),
     Input('sales-rep-filter', 'value'),
     Input('product-filter', 'value')]
)
def update_graphs(region, sales_rep, product):
    print(f"Filters: Region={region}, Sales Rep={sales_rep}, Product={product}")

    filtered_data = data.copy()

    # Apply region filter
    if region != 'All Regions':
        filtered_data = filtered_data[filtered_data['Region_ID'] == region]

    # Apply sales rep filter
    if sales_rep != 'All Sales Reps':
        filtered_data = filtered_data[filtered_data['Sales_Rep_ID'] == sales_rep]

    # Apply product filter (if not 'All Products', filter by product)
    if product != 'All Products':
        filtered_data = filtered_data[filtered_data['Product_ID'] == product]

    print(f"Filtered data shape: {filtered_data.shape}")

    # Handle the case where filtered_data might be empty
    if filtered_data.empty:
        return {}, {}, {}, {}, ["No data available for the selected filters."]

    # Sales by Day of Week
    sales_by_day = filtered_data.groupby('Day_of_Week')['Total_Sales'].sum().reset_index()
    sales_by_day_fig = px.bar(sales_by_day, x='Day_of_Week', y='Total_Sales', title="Sales by Day of the Week", text='Total_Sales')
    sales_by_day_fig.update_layout(template='plotly_dark')

    # Sales by Hour
    sales_by_hour = filtered_data.groupby('Hour')['Total_Sales'].sum().reset_index()
    sales_by_hour_fig = px.line(sales_by_hour, x='Hour', y='Total_Sales', markers=True, title="Sales by Hour")
    sales_by_hour_fig.update_layout(template='plotly_dark')

    # Profit vs Returns
    profit_vs_returns = filtered_data.groupby('Region_ID')[['Profit', 'Return_Rate']].sum().reset_index()
    profit_vs_returns_fig = px.bar(profit_vs_returns, x='Region_ID', y=['Profit', 'Return_Rate'], title="Profit vs Returns", barmode='group')
    profit_vs_returns_fig.update_layout(template='plotly_dark')

    # Product Performance (Sales by Product)
    sales_by_product = filtered_data.groupby('Product_ID')['Total_Sales'].sum().reset_index()
    sales_by_product_fig = px.bar(sales_by_product, x='Product_ID', y='Total_Sales', title="Sales by Product", text='Total_Sales')
    sales_by_product_fig.update_layout(template='plotly_dark')

    # Observations
    observations = []

    # Peak sales hour
    peak_hour = sales_by_hour.loc[sales_by_hour['Total_Sales'].idxmax()]
    observations.append(f"Peak sales occurred at {peak_hour['Hour']}:00 with {peak_hour['Total_Sales']:,} sales.")

    # Peak sales day
    peak_day = sales_by_day.loc[sales_by_day['Total_Sales'].idxmax()]
    observations.append(f"The highest sales occurred on {peak_day['Day_of_Week']} with {peak_day['Total_Sales']:,} sales.")

    # Return the updated figures and observations
    return sales_by_day_fig, sales_by_hour_fig, profit_vs_returns_fig, sales_by_product_fig, observations

if __name__ == '__main__':
    app.run_server(debug=True)
```
